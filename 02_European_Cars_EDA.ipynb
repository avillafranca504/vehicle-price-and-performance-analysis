## Factors Influencing Vehicle Prices and Performance in the European Automotive Market

Dataset: [European Car Dataset](https://www.kaggle.com/datasets/eswarpanchakarla/european-cars-dataset) | Source: Kaggle

### Dataset Context

This project uses the European Car Dataset, which, according to the dataset documentation, was compiled from European automotive sources, including manufacturer specifications, EV-database.org, and automotive market listings. The dataset contains information on 6,000 passenger vehicles available in the European market and includes 25 variables (features) describing vehicle characteristics.

The dataset covers multiple powertrain technologies, including electric vehicles (EVs), hybrids, petrol, and diesel models. The variables capture several dimensions of vehicle performance and characteristics, including technical specifications, efficiency measures, economic indicators, and safety-related information.

The dataset is provided in Excel (.xlsx) format, is licensed under CC BY 4.0, and is based on 2024 vehicle specifications.

### Feature Categories

The variables can be broadly grouped into the following categories:

- **Vehicle Identification & Classification:** Includes model name, brand (manufacturer), and classification information such as body type, market segment, and intended usage type.

- **Powertrain Characteristics:** Includes the propulsion system (powertrain) type (e.g., electric, hybrid, or internal combustion engine) and the vehicle's primary energy source (fuel type).

- **EV-Specific Specifications:** Includes usable battery capacity (kWh), estimated real-world driving range (km), energy efficiency (Wh/km), charging time (minutes), and maximum charging power (kW).

- **Performance Specifications:** Includes horsepower, torque (Nm), acceleration from 0–100 km/h (seconds), and top speed (km/h).

- **Practicality & Capacity:** Includes seating capacity, luggage compartment (boot) capacity (L), and towing capacity (kg).

- **Economic Indicators:** Includes vehicle purchase price (EUR), estimated annual maintenance costs (EUR), estimated energy or fuel cost per 100 km (EUR), and insurance rating.

- **Safety and Technology Features:** Includes the level of Advanced Driver Assistance Systems (ADAS) and Euro NCAP safety ratings.

### Limitations

The dataset was obtained from Kaggle and contains detailed specifications for vehicles available in the European market. Although the dataset documentation references multiple automotive sources, the complete data collection methodology and original source information are not fully disclosed by the publisher. Consequently, the results of this analysis should be interpreted as exploratory and may not be fully representative of the entire European vehicle market.

#### Data Validation

```python
# Investigate the segmentation classification
pd.crosstab(
    df['Fuel_Type'],
    df['Segment'])
```

```python
# The Segment variable combines two distinct classification systems.
# This feature distinguishes between size-based vehicle segments (A–F)
# and market-position categories used primarily for electric vehicles.

df["Segment_System"] = np.where(
    df["Segment"].isin(["A","B","C","D","E","F"]),
    "Vehicle_Size",
    "Market_Position"
)
```

**Data validation note**: The Segment variable contains two distinct classification schemes. Conventional and hybrid vehicles use the European A–F size-based segmentation, while electric vehicles use market-position categories (Economy, Mid-Range, Premium, Luxury, Performance). To make this distinction explicit, an additional variable (Segment_System) was created.

```python
# Investigate whether zero boot capacity values represent
# missing data or valid observations associated with specific vehicle types.

pd.crosstab(
    df["Boot_Capacity_(L)"] == 0,
    df["Body_Type"]
)
```

**Observation**: A review of the Boot_Capacity_(L) variable showed that all observations with a value of 0 corresponded to pickup vehicles. Since pickup trucks typically do not have a conventional enclosed luggage compartment, these values were retained and considered valid observations rather than missing data.

```python
# Insurance Rating Validation

# Price distribution by insurance rating
insurance_summary = (
    df.groupby("Insurance_Rating")
      .agg(
          Average_Price=("Price_(EUR)", "mean"),
          Vehicle_Count=("Price_(EUR)", "count")
      )
      .round(2)
)

display(insurance_summary)

# Correlation with target variable
corr = df[['Insurance_Rating', 'Price_(EUR)']].corr().iloc[0,1]
print(f"Correlation with Price: {corr:.3f}")

# Visual inspection
sns.boxplot(
    data=df,
    x='Insurance_Rating',
    y='Price_(EUR)'
)

plt.title("Price Distribution by Insurance Rating")
plt.show()

# Relationship with performance and safety
display(
    df.groupby('Insurance_Rating')[
        ['Horsepower_(HP)',
         'Price_(EUR)',
         'Safety_Rating_(Euro_NCAP)']
    ].mean().round(2)
)

# Correlation with performance and safety
corr = df[['Insurance_Rating', 'Horsepower_(HP)']].corr().iloc[0,1]
print(f"Correlation with Horsepower: {corr:.3f}")

corr = df[['Insurance_Rating', 'Safety_Rating_(Euro_NCAP)']].corr().iloc[0,1]
print(f"Correlation with Safety Rating: {corr:.3f}")
```

**Observations**: The dataset contains Insurance Rating values from 1 to 10 rather than the 1 to 5 range described in the documentation. Higher ratings are associated with higher vehicle prices and horsepower, while safety ratings remain relatively constant. Therefore, Insurance Rating appears to capture insurance risk associated with vehicle value and performance rather than safety alone.

```python
df['ADAS_Level'].value_counts().sort_index()
```

```python
# Standardize ADAS level values into a consistent ordinal scale
adas_mapping = {
    "0": 0,
    "1": 1,
    "2": 2,
    "3": 3,
    "L1": 1,
    "L2": 2,
    "L2+": 2.5,
    "L3": 3
}

df["ADAS_Level_Clean"] = df["ADAS_Level"].astype(str).map(adas_mapping)

print(df["ADAS_Level_Clean"].value_counts().sort_index())

# Validate the data type
print(df["ADAS_Level_Clean"].dtypes)
```

**Observations**: The original ADAS_Level variable contained mixed numeric and textual categories. These values were standardized into an ordinal variable, where higher values represent more advanced driver-assistance capability. The category L2+ was encoded as 2.5 to preserve its intermediate position between Level 2 and Level 3.
