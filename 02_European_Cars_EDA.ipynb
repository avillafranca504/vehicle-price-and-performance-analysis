```python
df.rename(columns={'Maintenance Cost (€/year)': 'Maintenance Cost (Euro/year)', 
                   'Energy Cost (€/100km)': 'Energy Cost (Euro/100km)'}, inplace=True) 

import re
df.rename(columns = lambda x:re.sub('[^A-Za-z0-9_/())]+', '_', x), inplace=True)

df.columns
```

```python
# Variable Type Categorization

# Target Variable

# target = 'Price_(EUR)'
# target = 'Segment'

#Categorical Variables

categorical_vars = [
    'Model_Name',
    'Brand',
    'Body_Type',
    'Segment',
    'Usage_Type',
    'Powertrain_Type',
    'Fuel_Type'
]

print("Categorical Variables:")
print("-" * 40)

for col in categorical_vars:
    print(f"{col}: {df[col].nunique()} unique values")

# Ordinal Variables (Variables with an inherent order/ranking)

ordinal_vars = [
    'ADAS_Level',
    'Safety_Rating_(Euro_NCAP)',
    'Insurance_Rating'
]

print("\nOrdinal Variables")
print("-" * 40)

for col in ordinal_vars:
    print(f"{col}: {df[col].nunique()} unique values")

# Continuous Numerical Variables (Measurements that can take decimal values)

continuous_vars = [
    'Price_(EUR)',
    'Maintenance_Cost_(Euro/year)',
    'Boot_Capacity_(L)',
    'Energy_Cost_(Euro/100km)',
    'Usable_Battery_(kWh)',
    'Real_Range_(km)',
    'Efficiency_(Wh/km)',
    '0_100_km/h_(s)',
    'Top_Speed_(km/h)',
    'Towing_Capacity_(kg)',
    'Max_Charging_Power_(kW)',
    'Horsepower_(HP)',
    'Torque_(Nm)', 
    'Charging_Time_(min)'
]

print("\nContinuous Numerical Variables")
print("-" * 40)

for col in continuous_vars:
    print(f"{col}: {df[col].nunique()} unique values")


# Discrete Numerical Variables (Countable integer values)

discrete_vars = ['Seating_Capacity']

print("\nDiscrete Numerical Variables")
print("-" * 40)

for col in discrete_vars:
    print(f"{col}: {df[col].nunique()} unique values")
```

```python
categorical_summary = df[categorical_vars].describe().T
categorical_summary
```

```python
ordinal_summary = df[ordinal_vars].astype('object').describe().T
ordinal_summary
```

```python
numerical_summary = df[continuous_vars + discrete_vars].describe().T
numerical_summary
```

```python
# Create a contingency table
stacked_data = pd.crosstab(df['Brand'], df['Powertrain_Type'])

# Plot stacked bar chart
stacked_data.plot(
    kind='bar',
    stacked=True,
    figsize=(16, 8)
)

plt.xlabel('Brand')
plt.ylabel('Count')
plt.title('Powertrain Type Distribution by Brand')
plt.legend(title='Powertrain Type')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```
